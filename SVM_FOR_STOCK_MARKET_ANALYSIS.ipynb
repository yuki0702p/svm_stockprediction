# Theory

**USING SUPPORT VECTOR MACHINE:**

supervised machine learning algorithm
classification, regression, and outlier detection tasks
The main goal of an SVM is to find the optimal hyperplane that best separates the data into different classes

KEY POINTS:

1. Hyperplane:

A hyperplane is a decision boundary that separates different classes in the feature space.
In a 2D space, a hyperplane is a line. In a 3D space, it is a plane, and in higher dimensions, it becomes a hyperplane.

2. Support Vectors:

Support vectors are the data points that are closest to the hyperplane. These points are crucial as they determine the position and orientation of the hyperplane.
The algorithm aims to maximize the margin, which is the distance between the hyperplane and the nearest support vectors of each class.

3. Margin:

The margin is the distance between the hyperplane and the closest support vectors. A larger margin implies better generalization of the classifier.
SVM tries to find the hyperplane that maximizes this margin.

Kernal:

1.Linear
2.Polynomial
3.Radial Basis Function
4.Sigmoid : tanh

**WHAT CODE DOES**

Using Support Vector Machines (SVM) for stock market analysis involves applying machine learning techniques to predict whether a stock should be bought or sold based on historical data and certain rules or features.

Problem Definition:

Classification Task: The primary task is to classify the stock's future movement into one of two classes: "Buy" or "Sell".

Derived Metrics: Percent changes, momentum, volatility measures, etc.

Kernel Selection: Choose an appropriate kernel (linear, polynomial, RBF) depending on the data distribution and complexity.



**PERFORMANCE MATRIX**
Accuracy
Definition: The proportion of correctly classified instances out of the total instances.

Formula:
Accuracy = TP+TN/TP+TN+FP+FN
​
Use Case: Simple to understand but can be misleading if the classes are imbalanced.

2. Precision
Definition: The proportion of true positive predictions out of all positive predictions made.

Formula:
Precision= TP/TP+FP
​
Use Case: Important when the cost of false positives is high.

3. Recall (Sensitivity or True Positive Rate)
Definition: The proportion of true positive predictions out of all actual positive instances.

Recall= TP/TP+FN

Use Case: Important when the cost of false negatives is high.

4. F1-Score
Definition: The harmonic mean of precision and recall, providing a balance between the two.

F1-Score=2×  Precision×Recall/Precision+Recall

Use Case: Useful when you need a balance between precision and recall.

5. Confusion Matrix
Definition: A table used to describe the performance of a classification model, showing the true positives (TP), true negatives (TN), false positives (FP), and false negatives (FN).

Use Case: Provides a comprehensive view of the model’s performance.

6. ROC Curve and AUC (Area Under the Curve)
ROC Curve: A plot of the true positive rate (recall) against the false positive rate (1-specificity) at various threshold settings.

AUC: The area under the ROC curve; a single scalar value that summarizes the performance.

Use Case: Evaluates the model's ability to distinguish between classes, useful
for imbalanced datasets.

7. Precision-Recall Curve
Definition: A plot of precision versus recall for different threshold values.
Use Case: Particularly useful for evaluating models on imbalanced datasets.

8. Specificity (True Negative Rate)
Definition: The proportion of true negative predictions out of all actual negative instances.

Specificity= TN/TN+FP

Use Case: Important when the cost of false positives is high.

**STANDARD SCALAR**

1. StandardScaler
Purpose: StandardScaler transforms data to have a mean of 0 and a standard deviation of 1. It is effective when the data follows a Gaussian distribution (normal distribution).

Formula:
z= x−μ/σ
where
x is the original feature value,
μ is the mean of the feature values, and
σ is the standard deviation of the feature values.

Use Cases:
Linear Models: Algorithms like Linear Regression, Logistic Regression, and SVM often benefit from standardized input data.

PCA (Principal Component Analysis): PCA is sensitive to the variance of features, making StandardScaler crucial for feature standardization.

Neural Networks: Standardizing inputs can help neural networks converge faster and reach better performance.

2. MinMaxScaler
Purpose: MinMaxScaler scales features to a fixed range, typically [0, 1], preserving the shape of the original distribution.

Formula:
xscaled= x−min(x)/max(x)−min(x)
where
x is the original feature value,
min(x) is the minimum value of x, and
max(x) is the maximum value of x.

Use Cases:

Neural Networks: Deep learning models often require inputs to be within a specific range for optimal performance.

Image Processing: Scaling pixel values of images to [0, 1] for neural network training.

Distance-based Algorithms: K-Means clustering and SVM with RBF kernels are sensitive to the scale of features.

3. Normalization
Purpose: Normalization scales each feature to a given range, often [0, 1], by dividing each feature by its magnitude in a way that the vector magnitude becomes 1.

Formula:
x normalized= x−min(x)/max(x)−min(x)
Use Cases:

Distance-based Algorithms: k-NN (k-Nearest Neighbors) requires normalized data as it computes distances between points.

Text Classification: Text features often require normalization to handle varying lengths and scales.

**GRID SEARCH**

Grid search is a technique used to find the optimal hyperparameters for a machine learning model.
When applying it to a Support Vector Machine (SVM) classifier, you typically consider parameters like

C(regularization parameter),

kernel (linear, polynomial, radial basis function (RBF)

gamma (kernel coefficient).

**KERNALS**

Kernals are used for defining the decision boundary and mapping input data into higher-dimensional feature spaces where linear separation is easier

1. Linear Kernel
Purpose: The linear kernel computes the dot product between two vectors, which is equivalent to no kernel (i.e., a linear decision boundary). It is suitable when the data is linearly separable.

Use Case:

When the number of features is large relative to the number of samples.
When the decision boundary is expected to be linear.

2. Polynomial Kernel
Purpose: The polynomial kernel calculates the similarity between two points in the feature space using polynomial functions of the original variables. It can handle non-linear decision boundaries.
Use Case:

When the data is moderately complex and non-linear.
Avoiding overfitting with lower degrees of polynomials.

3. Radial Basis Function (RBF) Kernel (Gaussian Kernel)
Purpose: The RBF kernel computes the similarity between two points based on the Euclidean distance between them in the original feature space. It is highly flexible and capable of capturing complex relationships in the data.

Use Case:

When there is no prior knowledge about the data or the data is expected to be non-linearly separable.
Handling complex decision boundaries.

4. Sigmoid Kernel
Purpose: The sigmoid kernel is based on the hyperbolic tangent function and can be seen as a two-layer neural network. It maps the input data into a higher-dimensional space.

Use Case:

When the data has no clear boundaries or is not linearly separable.
Non-linear mappings where higher-dimensional interactions between features are significant.

**INPUT VARIABLE**

```python
X = reliance[['Open','High','Low','Close']]
X.head()
```

```python
y = np.where(reliance['Close'].shift(1) < reliance['Close'],1,0)
```

**SPLIT THE DATA**

```python
split_percentage = 0.8
split = int(split_percentage*len(reliance))
# Train data set
X_train = X[:split]
y_train = y[:split]

# Test data set
X_test = X[split:]
y_test = y[split:]
```

**DATA PREPROCESSING**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**HYPERPARAMETER TUNING**

```python
param_grid = {
    'C': [0.1, 1, 10, 100],          # Regularization parameter
    'kernel': ['linear', 'rbf'],    # Kernel type
    'gamma': [0.1, 0.01, 0.001, 0.0001]  # Kernel coefficient for 'rbf' kernel
}
grid_search = GridSearchCV(estimator=SVC(), param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))

# Evaluate the best model on the test set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
print("Test set accuracy of best model: {:.2f}".format(test_accuracy))
```

**TRAIN THE MODEL**

```python
# Support Vector Machines
from sklearn.svm import SVC
lr = SVC(gamma=0.1, kernel='linear', C=100)
cls = lr.fit(X_train, y_train)
```

**TESTING**

```python
y_pred = lr.predict(X_test)
y_train_pred = lr.predict(X_train)
```

**EVALUATION METRICS**

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print('Confusion matrix: \n', confusion_matrix(y_test, y_pred))
print('Accuracy: ',accuracy_score(y_test, y_pred))
print('Accuracy_training: ',accuracy_score(y_train, y_train_pred))
```

**TRANFORM BACK TO ITS VALUES**

```python
testing_data = sc.inverse_transform(X_test)
```

```python
testing_data = pd.DataFrame(testing_data, columns=['Open','High','Low','Close'])
```

**CALCULATION**

```python
testing_data['predicted'] = y_pred
```
